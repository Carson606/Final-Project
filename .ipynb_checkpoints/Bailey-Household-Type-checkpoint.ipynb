# Pittsburgh's Best Neighborhood... For Burglars?

## Introduction

Using a dataset from the WPRDC, I have decided on the following metric to base my search for the best neighborhood on:

**Metric:** Nonfamily households 

We don't want to be unethical burglars stealing children's toys, do we? Aside from that, nonfamily households will likely be the most vulnerable, since the chance of someone occupying them at any given moment is lower. Nonfamily households also could potentially hold more high value items in them because the owners do not have to spend money on children.

### Households

```python
# imports
import pandas as pd
import csv
```

```python
# display a sample from the household types data 
types = pd.read_csv("household-type.csv")
types.head(10)
```

```python
# Bar graph of all neighborhoods and their number of lone householders

df = pd.DataFrame(types, columns = ['Neighborhood', 'Estimate; Nonfamily households: - Householder living alone']) 

# Creation of series of nonfamily, lone householders 
nfhouses = []
for ind in df.index: 
    nfhouses.append(df['Estimate; Nonfamily households: - Householder living alone'][ind])

# Creation of series of neighborhoods
hoods = []
for ind in df.index: 
    hoods.append(df['Neighborhood'][ind])

# Creation of bar graph & size adjustments   
pd.Series(nfhouses, index=hoods).plot.barh()

from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [20, 10]
%config InlineBackend.figure_format = 'retina'


# graph.set_xlabel("Crime per 100 Persons")
```

```python
# Bar graph of all neighborhoods and their proportion of lone householders vs total number of households
df = pd.DataFrame(types, columns = ['Neighborhood', 'Estimate; Nonfamily households: - Householder living alone', 'Estimate; Total:']) 

# Creation of series of the proportion of nonfamily, lone householders vs total number of households in a neighborhood
prop = []
for ind in df.index: 
    prop.append(df['Estimate; Nonfamily households: - Householder living alone'][ind]/df['Estimate; Total:'][ind])

# Creation of series of neighborhoods
hoods = []
for ind in df.index: 
    hoods.append(df['Neighborhood'][ind])

# Creation of bar graph & size adjustments
pd.Series(prop, index=hoods).plot.bar()
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [20, 10]
%config InlineBackend.figure_format = 'retina'
```
